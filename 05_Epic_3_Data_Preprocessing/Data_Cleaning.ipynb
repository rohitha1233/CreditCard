# Data Cleaning

The purpose of data cleaning is to fix structural anomalies, standardize values, and remove features that do not contribute to predictions. It is used to improve data quality and format variables for machine learning models. In this project, data cleaning is implemented by converting negative day markers, such as DAYS_BIRTH and DAYS_EMPLOYED, into positive years. We also drop the FLAG_MOBIL feature because it has zero variance, meaning all applicants have a registered mobile phone. This cleaning ensures variables represent actual years for age and employment. The result is a clean dataset with standardized columns, reducing feature dimensions and making values easier to interpret for both model training and deployment.


```python
import os
import pandas as pd
app_df = pd.read_csv(os.path.join('..', '03_Epic_1_Data_Collection', 'Dataset', 'application_record.csv'))
app_df['Age_Years'] = -app_df['DAYS_BIRTH'] / 365.25
app_df['Employed_Years'] = app_df['DAYS_EMPLOYED'].apply(lambda x: 0.0 if x > 0 else -x / 365.25)
app_df.drop(columns=['DAYS_BIRTH', 'DAYS_EMPLOYED', 'FLAG_MOBIL'], inplace=True)
print(app_df[['Age_Years', 'Employed_Years']].head())
```
