# Handling Missing Values

The purpose of handling missing values is to impute or remove null data entries within variables. It is used to ensure dataset completeness, as most machine learning algorithms cannot handle null values. In this project, missing value handling is implemented by scanning columns for null entries using the isnull().sum() function. The only column with missing values is OCCUPATION_TYPE, which has 1,436 null entries representing applicants who are unemployed or did not report their job. We impute these null values with the label 'Unknown' using fillna() to preserve applicant records. The result is a clean demographic dataset with zero missing entries, ensuring complete compatibility with subsequent encoders.


```python
import os
import pandas as pd
app_df = pd.read_csv(os.path.join('..', '03_Epic_1_Data_Collection', 'Dataset', 'application_record.csv'))
print(app_df.isnull().sum())
app_df['OCCUPATION_TYPE'] = app_df['OCCUPATION_TYPE'].fillna('Unknown')
print(f'Missing after imputation: {app_df["OCCUPATION_TYPE"].isnull().sum()}')
```
