# Drop Duplicate Records

The purpose of dropping duplicate records is to identify and remove redundant observations from the raw demographic and credit history datasets. This step is used to prevent data redundancy, which can cause model overfitting and inflate performance metrics. In this project, the duplicate removal is implemented using Pandas drop_duplicates() method. It is applied to both application_record.csv and credit_record.csv to clean the data before merging. The project-specific usage ensures that each applicant is represented by a single demographic profile. The result shows that there are no exact duplicate entries in either file, confirming that the raw files contain distinct applicant logs and are ready for downstream preprocessing steps.


```python
import os
import pandas as pd
app_df = pd.read_csv(os.path.join('..', '03_Epic_1_Data_Collection', 'Dataset', 'application_record.csv'))
print(f'Before drop: {len(app_df)}')
app_df.drop_duplicates(inplace=True)
print(f'After drop: {len(app_df)} ')
```
