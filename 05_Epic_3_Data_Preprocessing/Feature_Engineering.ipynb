# Feature Engineering

The purpose of feature engineering is to create new features that help machine learning models learn predictive patterns. It is used to capture relationships between variables, such as family dependencies and income groups. In this project, feature engineering is implemented by constructing a Family_Dependency ratio, which divides children count by total family size. We also bin the continuous annual income variable into categorical groups ('Low', 'Medium', 'High', 'Very High') based on quantiles. This project-specific usage helps the model capture family risk and income tiers. The result adds new features to the dataset, improving the model's ability to classify applicant risk profiles during training.


```python
import os
import pandas as pd
merged = pd.read_csv(os.path.join('..', '03_Epic_1_Data_Collection', 'Dataset', 'application_record.csv'))
merged['Family_Dependency'] = merged['CNT_CHILDREN'] / (merged['CNT_FAM_MEMBERS'] + 1e-5)
merged['Income_Category'] = pd.qcut(merged['AMT_INCOME_TOTAL'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
print(merged[['Family_Dependency', 'Income_Category']].head())
```
