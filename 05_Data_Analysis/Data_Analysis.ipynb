# Credit Card Approval Prediction - Epic 2: Visualizing and Analyzing Data

This notebook covers the detailed Exploratory Data Analysis (EDA) stage. We perform:
1. **Univariate Analysis**: Statistical distribution plots for individual features.
2. **Multivariate Analysis**: Evaluating relationships, correlations, and distributions between multiple features.
3. **Descriptive Analysis**: Key mathematical properties (means, standard deviations, skewness, kurtosis).

## 1. Import Core Libraries and Load Dataset

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Ensure outputs directory exists
out_dir = os.path.join('..', '11_Outputs')
os.makedirs(out_dir, exist_ok=True)

app_path = os.path.join('..', '03_Dataset', 'application_record.csv')
df = pd.read_csv(app_path)
print(f"Dataset loaded. Shape: {df.shape}")
```

## 2. Descriptive Analysis

We check the descriptive summary measures of the numerical columns. This includes the mean, median, mode, standard deviation, skewness, and kurtosis.

```python
print("--- Descriptive Summary Statistics ---")
print(df.describe().T)

print("\n--- Skewness ---")
print(df[['AMT_INCOME_TOTAL', 'CNT_CHILDREN', 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'CNT_FAM_MEMBERS']].skew())

print("\n--- Kurtosis ---")
print(df[['AMT_INCOME_TOTAL', 'CNT_CHILDREN', 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'CNT_FAM_MEMBERS']].kurt())

print("\n--- Modes ---")
print(df.mode().iloc[0])
```

## 3. Univariate Analysis

We create distribution plots for categorical and numerical features. We save these graphs in the outputs folder.

```python
sns.set_theme(style="whitegrid")

# 3.1 Annual Income Distribution (Histogram)
plt.figure(figsize=(10, 5))
sns.histplot(df['AMT_INCOME_TOTAL'], bins=30, kde=True, color='#2563EB')
plt.title('Annual Income Distribution', fontsize=14, fontweight='bold', color='#1E3A8A')
plt.xlabel('Income (units)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.savefig(os.path.join(out_dir, 'histogram.png'), dpi=150, bbox_inches='tight')
plt.show()

# 3.2 Occupation Type Distribution (Count Plot)
plt.figure(figsize=(12, 6))
sns.countplot(y='OCCUPATION_TYPE', data=df, order=df['OCCUPATION_TYPE'].value_counts().index, palette='viridis')
plt.title('Occupation Type Distribution', fontsize=14, fontweight='bold', color='#1E3A8A')
plt.xlabel('Count', fontsize=12)
plt.ylabel('Occupation', fontsize=12)
plt.savefig(os.path.join(out_dir, 'countplot.png'), dpi=150, bbox_inches='tight')
plt.show()

# 3.3 Pie Chart of Income Type
plt.figure(figsize=(8, 8))
df['NAME_INCOME_TYPE'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90)
plt.title('Income Type Share', fontsize=14, fontweight='bold', color='#1E3A8A')
plt.ylabel('')
plt.savefig(os.path.join(out_dir, 'income_type_pie.png'), dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Multivariate Analysis

We investigate relationships between features.

```python
# 4.1 Income vs Education (Boxplot)
plt.figure(figsize=(12, 6))
sns.boxplot(x='AMT_INCOME_TOTAL', y='NAME_EDUCATION_TYPE', data=df, palette='Blues')
plt.title('Income Distribution across Education Levels', fontsize=14, fontweight='bold', color='#1E3A8A')
plt.xlabel('Income', fontsize=12)
plt.ylabel('Education Type', fontsize=12)
plt.xscale('log') # Scale because of outliers
plt.savefig(os.path.join(out_dir, 'boxplot.png'), dpi=150, bbox_inches='tight')
plt.show()

# 4.2 Age vs Employment (Scatter plot)
# Convert birth and employment days to positive years
df['Age_Years'] = -df['DAYS_BIRTH'] / 365.25
df['Employed_Years'] = df['DAYS_EMPLOYED'].apply(lambda x: 0 if x > 0 else -x / 365.25)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age_Years', y='Employed_Years', hue='NAME_INCOME_TYPE', data=df, alpha=0.6, palette='Set2')
plt.title('Age vs Employment Years', fontsize=14, fontweight='bold', color='#1E3A8A')
plt.xlabel('Age (Years)', fontsize=12)
plt.ylabel('Employed (Years)', fontsize=12)
plt.savefig(os.path.join(out_dir, 'age_vs_employment.png'), dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Correlation Analysis

```python
# 5.1 Correlation Heatmap of Numerical Features
plt.figure(figsize=(10, 8))
corr = df[['CNT_CHILDREN', 'AMT_INCOME_TOTAL', 'Age_Years', 'Employed_Years', 'CNT_FAM_MEMBERS']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap', fontsize=14, fontweight='bold', color='#1E3A8A')
plt.savefig(os.path.join(out_dir, 'correlation_heatmap.png'), dpi=150, bbox_inches='tight')
plt.savefig(os.path.join(out_dir, 'heatmap.png'), dpi=150, bbox_inches='tight')
plt.show()

# 5.2 Pairplot of Numerical columns
plt.figure(figsize=(12, 12))
pair_grid = sns.pairplot(df[['AMT_INCOME_TOTAL', 'Age_Years', 'Employed_Years', 'CNT_FAM_MEMBERS']], diag_kind='kde', plot_kws={'alpha':0.4})
pair_grid.fig.suptitle('Pairplot of Numerical Features', y=1.02, fontsize=14, fontweight='bold', color='#1E3A8A')
pair_grid.savefig(os.path.join(out_dir, 'pairplot.png'), dpi=150, bbox_inches='tight')
plt.show()
```
