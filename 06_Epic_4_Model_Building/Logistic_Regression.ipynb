### Introduction:
Logistic Regression is a linear classification model that predicts binary outcomes. It is used as a baseline model due to its simplicity, ease of training, and high explainability. In this project, the model is trained on an 80/20 train/test stratified split of the scaled demographic features. Training is implemented using Scikit-Learn's LogisticRegression class with the L-BFGS solver. We evaluate the model using Accuracy, Precision, Recall, F1 Score, and ROC-AUC. The project results show that Logistic Regression achieves a test accuracy of 82.30% and a recall of 100.00% on the default class. This high recall indicates that the model classifies all high-risk accounts correctly, but its precision is limited because it flags many low-risk accounts as high-risk.
