# Decision Tree

### Objective:
Train and evaluate a Decision Tree classifier.

### Introduction:
The Decision Tree Classifier is a non-linear model that splits data based on feature thresholds. It is used because it can capture non-linear relationships and interactions between features without requiring normal distributions. In this project, the model is trained on the stratified training set with a maximum depth of 6 to prevent overfitting. Training is implemented using Scikit-Learn's DecisionTreeClassifier class. The model is evaluated on the test set using classification matrices and confusion heatmaps. The project results show that the Decision Tree achieves an accuracy of 81.60%, a precision of 82.24%, and a recall of 99.03%. The decision boundaries split the data based on employment duration and age, showing that younger applicants with short employment histories have a higher risk of default.
