# Random Forest

### Objective:
Train a Random Forest ensemble model to predict applicant default risk.

### Introduction:
Random Forest is an ensemble bagging classifier that aggregates predictions from multiple decision trees. It is used to reduce variance, improve generalization, and prevent overfitting by training trees on random feature subsets. In this project, the forest is trained using 150 estimators with a maximum depth of 8. Training is implemented using Scikit-Learn's RandomForestClassifier class. The model is evaluated on the test set, and feature importances are extracted to identify key predictors. The project results show that Random Forest achieves a test accuracy of 82.30% and a recall of 100.00%. The feature importance plot reveals that employment duration, age, and annual income are the most significant variables in predicting credit approval, matching our expectations.
