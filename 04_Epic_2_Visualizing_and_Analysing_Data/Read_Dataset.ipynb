# Reading Dataset

This notebook handles reading the raw demographic and credit history datasets. It uses Pandas to load application_record.csv and credit_record.csv from the collection folder. Loading the datasets is a crucial step to inspect shapes, columns, schema details, and row heads before performing analysis. This step checks data dimensions, verifies ID alignments, and counts unique records in both files. In this project, the data is loaded using relative filepaths, which confirms path validations. Identifying the structural formats of both records at this point ensures the preprocessing steps are aligned correctly with the database schemas.


```python
import os
import pandas as pd
app_path = os.path.join('..', '03_Epic_1_Data_Collection', 'Dataset', 'application_record.csv')
credit_path = os.path.join('..', '03_Epic_1_Data_Collection', 'Dataset', 'credit_record.csv')
app_df = pd.read_csv(app_path)
credit_df = pd.read_csv(credit_path)
print(f'Application records shape: {app_df.shape}')
print(f'Credit histories shape: {credit_df.shape}')
```
