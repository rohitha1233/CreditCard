# Bivariate Association Analysis

### Objective:
Explore relationships between two variables to capture predictive insights.

### Introduction:
Bivariate analysis evaluates the relationships and interactions between two distinct variables. It is used to identify how individual predictors interact with each other and with the target label before model training. In this project, bivariate analysis is implemented by generating box plots, scatter plots, and grouped bar charts. We analyze income distributions across education levels (box plots), plot relationships between age and annual income (scatter plots), and examine how gender splits across credit status target categories (grouped bar charts). The resulting figures are saved as PNG files. These plots show that income scale rises with higher education levels, and age splits contain zero collinearity with income, providing distinct predictors.


```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
app_df = pd.read_csv(os.path.join('..', '03_Epic_1_Data_Collection', 'Dataset', 'application_record.csv'))
app_df['Age_Years'] = -app_df['DAYS_BIRTH'] / 365.25
# Box Plot of Income by Education
plt.figure(figsize=(10,5))
sns.boxplot(x='AMT_INCOME_TOTAL', y='NAME_EDUCATION_TYPE', data=app_df)
plt.xscale('log')
plt.title('Income Distribution by Education')
plt.tight_layout()
plt.close()
```

```python
# Scatter plot of Age vs Income
plt.figure(figsize=(8,5.5))
sns.scatterplot(x=app_df['Age_Years'], y=app_df['AMT_INCOME_TOTAL'], alpha=0.5)
plt.title('Age vs Annual Income')
plt.savefig(os.path.join('Outputs', 'ScatterPlot.png'))
plt.close()
```

### Observations & Conclusion:
The box plot shows that academic degree and higher education levels yield higher median annual incomes. The scatter plot reveals that age and income are weakly correlated, indicating that age does not dictate income scale in this cohort.
