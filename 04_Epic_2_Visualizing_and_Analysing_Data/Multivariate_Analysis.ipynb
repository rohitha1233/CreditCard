# Multivariate Feature Interaction Analysis

### Objective:
Map multicollinearities and multi-feature associations using matrices and heatmaps.

### Introduction:
Multivariate analysis explores the relationships, correlations, and interactions between two or more variables in a dataset. It is used to detect correlations, check multicollinearity, and identify features that are strongly related to applicant credit risk. In this project, multivariate analysis is implemented by creating a correlation matrix heatmap, boxplots comparing income across education levels, and pairplots. The resulting charts are saved as PNG files in the outputs directory. The correlation heatmap reveals a positive correlation between family size and children count, while the boxplot confirms that higher education levels are associated with higher annual incomes.


```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
app_df = pd.read_csv(os.path.join('..', '03_Epic_1_Data_Collection', 'Dataset', 'application_record.csv'))
app_df['Age_Years'] = -app_df['DAYS_BIRTH'] / 365.25
app_df['Employed_Years'] = app_df['DAYS_EMPLOYED'].apply(lambda x: 0.0 if x > 0 else -x / 365.25)
plt.figure(figsize=(8, 6.5))
corr = app_df[['CNT_CHILDREN', 'AMT_INCOME_TOTAL', 'Age_Years', 'Employed_Years', 'CNT_FAM_MEMBERS']].corr()
sns.heatmap(corr, annot=True, fmt='.3f', cmap='coolwarm')
plt.savefig(os.path.join('Outputs', 'Correlation_Heatmap.png'))
plt.close()
print('Plot saved.')
```

### Observations & Conclusion:
The correlation matrix confirms that family size and children count are highly correlated (0.88), which is expected. Other numerical variables show weak correlations (below 0.1), indicating low multicollinearity and verifying that features represent distinct information.
