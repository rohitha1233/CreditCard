# Univariate Data Visualization

### Objective:
Plot individual feature distributions and frequency counts of categorical columns.

### Introduction:
Univariate analysis examines the distribution of a single feature in isolation. It is used to identify distribution shapes, target frequencies, and categorical breakdowns. In this project, univariate analysis is implemented by generating seaborn distribution plots and countplots. We visualize the distributions of annual income, gender ratios, and family member counts. The plots are saved as PNG files in the outputs folder. The visual outputs show that the applicant population is dominated by female candidates and is concentrated in the middle-income range. This visual check highlights structural patterns that could cause class imbalances or biases in model predictions.


```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
app_df = pd.read_csv(os.path.join('..', '03_Epic_1_Data_Collection', 'Dataset', 'application_record.csv'))
plt.figure(figsize=(8,5))
sns.histplot(app_df['AMT_INCOME_TOTAL'], bins=30, kde=True)
plt.title('Income Distribution')
plt.savefig(os.path.join('Outputs', 'Histogram.png'))
plt.close()
print('Plot saved.')
```

```python
# Countplot for CODE_GENDER
sns.countplot(x='CODE_GENDER', data=app_df)
plt.title('Gender Count distribution')
plt.close()
```

### Observations & Conclusion:
The income distribution has a clean log-normal pattern. Categorical counts show that about two-thirds of the applicants are female, and property owners exceed car owners, highlighting demographic trends in the applicant cohort.
