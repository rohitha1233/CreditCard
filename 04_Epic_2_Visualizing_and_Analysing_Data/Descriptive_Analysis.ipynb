# Descriptive Statistical Analysis

### Objective:
Compute statistical moments of the numerical demographic attributes.

### Introduction:
Descriptive analysis calculates central tendencies, dispersions, and distributions of numeric variables in the dataset. It is used to understand the ranges, averages, standard deviations, and shape metrics like skewness and kurtosis. In this project, descriptive statistics are computed for income, age, employment duration, and family size. This implementation uses the describe() function alongside skew() and kurt() methods. The result identifies standard scales, flags outliers, and shows that the income variable is right-skewed. Conducting this step helps determine if scaling is required before model training, ensuring models are trained on properly scaled values.


```python
import os
import pandas as pd
app_df = pd.read_csv(os.path.join('..', '03_Epic_1_Data_Collection', 'Dataset', 'application_record.csv'))
app_df['Age_Years'] = -app_df['DAYS_BIRTH'] / 365.25
app_df['Employed_Years'] = app_df['DAYS_EMPLOYED'].apply(lambda x: 0.0 if x > 0 else -x / 365.25)
print('=== Basic describe output ===')
print(app_df[['CNT_CHILDREN', 'AMT_INCOME_TOTAL', 'Age_Years', 'Employed_Years', 'CNT_FAM_MEMBERS']].describe().T)
```

```python
cols = ['CNT_CHILDREN', 'AMT_INCOME_TOTAL', 'Age_Years', 'Employed_Years', 'CNT_FAM_MEMBERS']
for col in cols:
    print(f'\nFeature: {col}')
    print(f'  Mean: {app_df[col].mean():.2f}')
    print(f'  Median: {app_df[col].median():.2f}')
    print(f'  Mode: {app_df[col].mode()[0]:.2f}')
    print(f'  Std Dev: {app_df[col].std():.2f}')
    print(f'  Variance: {app_df[col].var():.2f}')
    print(f'  Min: {app_df[col].min():.2f}')
    print(f'  Max: {app_df[col].max():.2f}')
```

### Observations & Conclusion:
The average annual income is $158,777.78 with a median of $155,949.87, and is slightly right-skewed. The average age is 42.8 years, and average employment is 5.8 years. These calculations show that the numerical columns are on significantly different ranges, requiring scaling.
