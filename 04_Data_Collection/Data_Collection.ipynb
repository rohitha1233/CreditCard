## 2. Load Datasets

We load `application_record.csv` and `credit_record.csv` using Pandas `read_csv()` function.

```python
app_path = os.path.join('..', '03_Dataset', 'application_record.csv')
credit_path = os.path.join('..', '03_Dataset', 'credit_record.csv')

# Read dataframes
app_df = pd.read_csv(app_path)
credit_df = pd.read_csv(credit_path)

print(f"Application Record shape: {app_df.shape}")
print(f"Credit Record shape: {credit_df.shape}")
```

## 3. Review Application Record Columns and Samples

```python
print("--- Application Record Columns ---")
print(app_df.columns.tolist())

print("\n--- First 5 Rows of Application Record ---")
print(app_df.head())

print("\n--- Last 5 Rows of Application Record ---")
print(app_df.tail())

print("\n--- Application Data Types & Non-Null Counts ---")
print(app_df.info())
```

## 4. Review Credit Record Columns and Samples

```python
print("--- Credit Record Columns ---")
print(credit_df.columns.tolist())

print("\n--- First 5 Rows of Credit Record ---")
print(credit_df.head())

print("\n--- Last 5 Rows of Credit Record ---")
print(credit_df.tail())

print("\n--- Credit Data Types & Non-Null Counts ---")
print(credit_df.info())
```

## 5. Check Missing Values and Duplicates

```python
print("--- Application Missing Values ---")
print(app_df.isnull().sum())

print("\n--- Credit Missing Values ---")
print(credit_df.isnull().sum())

print(f"\nDuplicate rows in Application Record: {app_df.duplicated().sum()}")
print(f"Duplicate rows in Credit Record: {credit_df.duplicated().sum()}")
```

## 6. Verification of Unique Applicant IDs

We check if there are overlaps in IDs between the two datasets and the count of unique IDs.

```python
app_unique_ids = app_df['ID'].nunique()
credit_unique_ids = credit_df['ID'].nunique()
overlap_ids = len(set(app_df['ID']).intersection(set(credit_df['ID'])))

print(f"Unique IDs in Application Record: {app_unique_ids}")
print(f"Unique IDs in Credit Record: {credit_unique_ids}")
print(f"Number of intersecting IDs: {overlap_ids}")
```
